```python
x_train= x_train.drop("loan_id", axis = 1)
x_test= x_test.drop("loan_id", axis = 1)

# loan id is an just identifier which has no meaning, if we k
# it, the model may learn noise
# removing unneccesary column from train and test model
```

```python
# encoding the model x train and x test
# one hot encode categorical features

from sklearn.preprocessing import OneHotEncoder

categorical_cols = ["education","self_employed"]
ohe =OneHotEncoder(drop='first',sparse_output=False)
x_train_cat = ohe.fit_transform(x_train[categorical_cols])
x_test_cat = ohe.transform(x_test[categorical_cols])

# the drop is to remove reduntant columns to avoid dummy
# variable trap
# sparse is default would store them as numpy array,
# if false itwill give the
```

```python
# scalling numeric features

from sklearn.preprocessing import StandardScaler
numeric_cols = x_train.drop(columns=categorical_cols).columns
scaler = StandardScaler()
x_train_num=scaler.fit_transform(x_train[numeric_cols])
x_test_num=scaler.transform(x_test[numeric_cols])
```

```python
x_train_num_df = pd.DataFrame(
    x_train_num,
    columns=numeric_cols,
    index = x_train.index
)

x_train_num_df.head()
```

```python

x_train_final = np.hstack([x_train_num,x_train_cat])

x_test_final = np.hstack([x_test_num,x_test_cat])

```

```python
# here logistic regression is the correct one
# because the target is categorical output - approved/not approved
# logistic regression is used for classification and predict probability of process
# converts it into class0 or 1
# its simple , easy & fast to implement
```

```python
from collections import UserString
# logistic regression model

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(x_train_final,y_train)


# max_iter=1000 is for LogisticRegression to learnt UserString
# the number of iterations(optimization)
# default iterations are not enough model fails to prove
# this is where the model learns the pattern from the data
```

```python
# evaluate accuracy

from sklearn.metrics import accuracy_score
y_pred = model.predict(x_test_final)
accuracy_score(y_test,y_pred)
```
